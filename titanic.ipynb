```python
def fillage(cols):
    age=cols[0]
    pclass=cols[1]
    
    if (pd.isnull(age)):
        if (pclass==1):
            return 38
        if (pclass==2):
            return 30
        if (pclass==3):
            return 25
    else:
        return age
```

```python
df['Age']=df[['Age','Pclass']].apply(fillage,axis=1)
```

```python
df.Age
```

```python
sns.heatmap(df.isnull())
```

```python
df.drop('Cabin',axis=1,inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
df.info()
```

```python
df.info()
```

```python
df.drop(['PassengerId','Ticket','Fare'],axis=1,inplace=True)
```

```python
df.drop(['Name'],axis=1,inplace=True)
```

```python
df.head()
```

```python
from sklearn.preprocessing import OrdinalEncoder
oe=OrdinalEncoder()
df[['Sex','Embarked']]=oe.fit_transform(df[['Sex','Embarked']])
```

```python
X=df.drop('Survived',axis=1)
```

```python
X
```

```python
Y=df['Survived']
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(X,Y,test_size=0.3,random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(xtrain,ytrain)
ypred=lr.predict(xtest)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
cm=confusion_matrix(ytest,ypred)
cr=classification_report(ytest,ypred)
ac=accuracy_score(ytest,ypred)
print(f"Accuracy score: {ac}\n{cm}\n {cr}" )
```

```python
train=lr.score(xtrain,ytrain)
test=lr.score(xtest,ytest)
print(f"Training accuracy: {train}\n Testing accuracy:{test}")
```

```python
# As testing score > training score by 5% we have scenario of
# high bias & low variance- underfitting
```
